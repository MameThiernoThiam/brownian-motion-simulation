```python
import numpy as np
import matplotlib.pyplot as plt

def brownian_motion(T=1, N=1000):
    dt = T / N
    increments = np.sqrt(dt) * np.random.randn(N)  # Loi normale N(0, dt)
    W = np.cumsum(increments)  # Somme cumule
    W = np.insert(W, 0, 0)  # On ajoute W_0 = 0
    return W

if __name__ == "__main__":
    T = 1
    N = 1000
    W = brownian_motion(T, N)

    t = np.linspace(0, T, N+1)
    plt.plot(t, W)
    plt.title("Mouvement Brownien 1D")
    plt.xlabel("Temps")
    plt.ylabel("W(t)")
    plt.grid(True)
    plt.show()
if __name__ == "__main__":
    T = 1
    N = 1000
    nb_paths = 10  # nombre de trajectoires

    t = np.linspace(0, T, N+1)

    for _ in range(nb_paths):
        W = brownian_motion(T, N)
        plt.plot(t, W, linewidth=1)

    plt.title("10 Trajectoires de Mouvement Brownien 1D")
    plt.xlabel("Temps")
    plt.ylabel("W(t)")
    plt.grid(True)
    plt.show()

```
